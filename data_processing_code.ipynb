```python
lst_years = [2015, 2016, 2017, 2018, 2019, 2020]

str_directory_linux = '/media/tommy/Data/IAT/xlsx_to_csv/'
str_directory_windows = "D:/IAT/xlsx_to_csv/"
lst_directory = [str_directory_windows, str_directory_linux]

li = []

for year in lst_years:
    str_file = 'property-tax-report_' + str(year) + '.csv'

    df = pd.read_csv(lst_directory[int_environment] + str_file, usecols=None, nrows=None, dtype={'NEIGHBOURHOOD_CODE': str, 'REPORT_YEAR': int})
    li.append(df)
    
df_all = pd.concat(li, axis=0, ignore_index=False)

df_all.head()
```

```python
df2 = df_all.copy()

# Filter by Zone Category
lst_zone_categories = ['One Family Dwelling', 'Multiple Family Dwelling', 'Comprehensive Development', 'Two Family Dwelling']
df2 = df2[df2['ZONE_CATEGORY'].isin(lst_zone_categories)]

df2['dwelling_type'] = np.where(
    df2['ZONE_CATEGORY']=='Comprehensive Development',
    'condo_dwelling',
    'family_dwelling'
)

# Filter out OTHER

df2 = df2[df2['LEGAL_TYPE']!='OTHER']
# Remove all Zero Taxes
# df2 = df2[df2['TAX_LEVY'] > 0]
```

```python
df2.head()
```

```python
# REMOVE UNNECESSARY COLUMNS
lst_columns = ['PID', 'LEGAL_TYPE', 'ZONE_CATEGORY', 'CURRENT_LAND_VALUE', 'CURRENT_IMPROVEMENT_VALUE', 'PREVIOUS_IMPROVEMENT_VALUE', 'PREVIOUS_LAND_VALUE', 'TAX_LEVY', 'NEIGHBOURHOOD_CODE', 'REPORT_YEAR', 'dwelling_type', 'YEAR_BUILT']
lst_address = ['TO_CIVIC_NUMBER', 'STREET_NAME', 'PROPERTY_POSTAL_CODE']
df2 = df2[lst_columns + lst_address]
```

```python
# Sum Total Value of Current Year's Land and Building
df2['CURRENT_TOTAL_VALUE'] = df2['CURRENT_LAND_VALUE'].fillna(0) + df2['CURRENT_IMPROVEMENT_VALUE'].fillna(0)
```

```python
# Removing Records with no CURRENT_VALUE
df2 = df2[df2['CURRENT_TOTAL_VALUE']>0]
```

```python
# Sum Total Value of Last Year's Land and Building
df2['PREVIOUS_TOTAL_VALUE'] = df2['PREVIOUS_LAND_VALUE'].fillna(0) + df2['PREVIOUS_IMPROVEMENT_VALUE'].fillna(0)
```

```python
# Change in Value
df2['CHANGE_TOTAL_VALUE'] = df2['CURRENT_TOTAL_VALUE'] - df2['PREVIOUS_TOTAL_VALUE']
df2['mortgage_return_prctg'] = (df2['CHANGE_TOTAL_VALUE'] / df2['PREVIOUS_TOTAL_VALUE'] ) * 100
```

```python
# Add number of bedrooms NEW
# https://www.zolo.ca/vancouver-real-estate/trends

lst_years = [2014, 2015, 2016, 2017, 2018, 2019]#, 2020]
# lst_years = [2015]

lst_dwelling_type = ['condo_dwelling', 'family_dwelling']

li = []

for year in lst_years:
    for dwelling_type in lst_dwelling_type:
#         print(year, dwelling_type)
        
        str_directory_windows = 'C:/Users/tommy/Dropbox/SFU/IAT 814/Project/data/MLS_property_sales/'
        str_directory_linux = '/home/tommy/Dropbox/SFU/IAT 814/Project/data/MLS_property_sales/'
        lst_directory = [str_directory_windows, str_directory_linux]
        str_file = 'average_sale_prices_bedroom_dwelling_copied_2020.xlsx'
        lst_columns = ['year', 'dwelling_type', 'bedrooms', 'range_start', 'range_end']
        str_sheet = 'average_sale_prices'
        average_sale_prices_bedroom_dwelling = pd.read_excel(lst_directory[int_environment] + str_file, sheet_name=str_sheet, usecols=lst_columns, nrows=None, dtype={'year': np.int32, 'range_start': np.float64, 'range_end': np.float64})

        average_sale_prices_bedroom_dwelling = average_sale_prices_bedroom_dwelling[average_sale_prices_bedroom_dwelling['year']==year]
        average_sale_prices_bedroom_dwelling = average_sale_prices_bedroom_dwelling[average_sale_prices_bedroom_dwelling['dwelling_type']==dwelling_type]
        
        df2_temp = df2.copy()
        df2_temp = df2_temp[df2_temp['REPORT_YEAR']==year]
        df2_temp = df2_temp[df2_temp['dwelling_type']==dwelling_type]

        average_sale_prices_bedroom_dwelling.index = pd.IntervalIndex.from_arrays(average_sale_prices_bedroom_dwelling['range_start'],average_sale_prices_bedroom_dwelling['range_end'],closed='both')
        df2_temp['bedrooms'] = df2_temp['CURRENT_TOTAL_VALUE'].apply(lambda x : average_sale_prices_bedroom_dwelling.iloc[average_sale_prices_bedroom_dwelling.index.get_loc(x)]['bedrooms'])
        
        li.append(df2_temp)
    
df3 = pd.concat(li, axis=0, ignore_index=False)
df3.reset_index(drop=True, inplace=True)
df3['bedrooms'] = df3['bedrooms'].astype(np.int64)
```

```python
# Remove Condows with values too low or too high for residential purposes

str_directory_windows = 'C:/Users/tommy/Dropbox/SFU/IAT 814/Project/data/MLS_property_sales/'
str_directory_linux = '/home/tommy/Dropbox/SFU/IAT 814/Project/data/MLS_property_sales/'
lst_directory = [str_directory_windows, str_directory_linux]
str_file = 'average_sale_prices_bedroom_dwelling.xlsx'
lst_columns = ['year', 'dwelling_type', 'bedrooms', 'range_start_valid', 'range_end_valid']
str_sheet = 'average_sale_prices'
average_sale_prices_bedroom_dwelling = pd.read_excel(lst_directory[int_environment] + str_file, sheet_name=str_sheet, usecols=lst_columns, nrows=None, dtype={'year': np.int32})

df3 = pd.merge(df3, average_sale_prices_bedroom_dwelling,  how='left', left_on=['REPORT_YEAR', 'bedrooms', 'dwelling_type'], right_on = ['year', 'bedrooms', 'dwelling_type'])

df3['invalid_value'] = np.nan
# df3['invalid_value'] = np.where(
#     ((df3['dwelling_type'] == 'condo_dwelling') & ((df3['CURRENT_TOTAL_VALUE'] <= df3['range_start_valid']) | (df3['CURRENT_TOTAL_VALUE'] >= df3['range_end_valid']))),
#     'TRUE',
#     df3['invalid_value']                                                                                                                      
# )

df3['invalid_value'] = np.where(
    (((df3['CURRENT_TOTAL_VALUE'] <= df3['range_start_valid']) | (df3['CURRENT_TOTAL_VALUE'] >= df3['range_end_valid']))),
    'TRUE',
    df3['invalid_value']                                                                                                                      
)

df3 = df3[df3['invalid_value']!='TRUE']
df3.drop(columns=({'year', 'range_start_valid', 'range_end_valid', 'invalid_value'}), inplace=True)

```

```python
# ADD Maintenance Fees
float_maintenance_fee_prct = 0.04 # old 0.002
df3['maintenance_fee'] = np.nan
df3['maintenance_fee'] = np.where(
    df3['LEGAL_TYPE'] == 'STRATA',
    df3['CURRENT_TOTAL_VALUE'] * float_maintenance_fee_prct/2,
    df3['CURRENT_TOTAL_VALUE'] * float_maintenance_fee_prct
)
```

```python
# ADD Strata Fees
float_strata_fees_prctg = 0.005013135

df3['strata_fee'] = np.where(
    ((df3['LEGAL_TYPE']== 'STRATA')),
    df3['CURRENT_TOTAL_VALUE'] * float_strata_fees_prctg,
    0
    )
```

```python
# ADD Homeowners Insurance
df3['homeowners_insurance'] = np.nan

lst_homeowners_insurance_thresholds = [0, 100000, 300000, 700000, 1500000, 5000000]

homeowners_insurance_dict = {
            0: 43,
            100000: 51,
            300000: 87,
            700000: 130,
            1500000: 155,
            5000000: 300,
        }
#https://insureye.com/average-home-insurance-cost-overpaying-do-a-30-second-test/


for int_threshold in lst_homeowners_insurance_thresholds:
    
    df3['homeowners_insurance'] = np.where(
    df3["CURRENT_TOTAL_VALUE"] >= int_threshold,
    homeowners_insurance_dict[int_threshold],
    df3['homeowners_insurance']
    )
    
df3['homeowners_insurance'] = 12 * df3['homeowners_insurance']
```

```python
# Add Total Fees
df3['total_owners_fees'] = df3.maintenance_fee + df3.strata_fee + df3.homeowners_insurance
```

```python
# ALL THE CALCULATIONS


df5 = df4.copy()

lst_dimensions = ['year_start', 'year_end', 'down_payment', 'Mortgage_Fixed_Term', 'Loan_Life', 'stock', 'LEGAL_TYPE', 'ZONE_CATEGORY', 'dwelling_type', 'REPORT_YEAR', 'bedrooms', 'Neighbourhood_Name', 'total_renters_cost', 'value_at_purchase', 'loan_amount_at_mortgage_contract', 'mortgage_investment_value', 'investment_at_purchase', 'mortgage_return_prctg_yearly', 'mortgage_return_prctg_cumulative', 'stocks_w_margin_investment_value_after_rental_costs', 'return_prctg_yearly_stocks_w_margin', 'return_prctg_cumulative_stocks_w_margin', 'stocks_wo_margin_investment_value_after_rental_costs', 'return_prctg_yearly_stocks_wo_margin', 'return_prctg_cumulative_stocks_wo_margin']
lst_measures = ['CURRENT_TOTAL_VALUE_CORRECTED', 'PREVIOUS_TOTAL_VALUE_CORRECTED', 'mortgage_return_prctg', 'total_owners_fees', 'owning_costs_total_yearly', 'mortgage_return_value_yearly', 'mortgage_return_prctg_annualized', 'return_prctg_cumulative_stocks_w_margin_annualized', 'return_prctg_cumulative_stocks_wo_margin_annualized']
lst_columns = lst_dimensions + lst_measures
str_directory_windows = "D:/IAT/tableau_input/"
str_directory_linux = '/home/tommy/Documents/IAT/project/tableau_input/'
lst_directory = [str_directory_windows, str_directory_linux]

df5 = df5.sort_values(['LEGAL_TYPE', 'ZONE_CATEGORY', 'bedrooms', 'Neighbourhood_Name', 'REPORT_YEAR'], ascending=[True, True, True, True, True])
df5.reset_index(drop=True, inplace=True) 

# Mortgage Calculations Parameters
float_opening_costs = 0.01
float_closing_costs = 0.05

lst_down_payments = [.05, .1, .2, .3, .5, .999999]
# lst_down_payments = [.05, .2, .3, .5, .9999]
# lst_down_payments = [.05, .1, .2]

# lst_down_payments = [.9999]
# lst_down_payments = [.2]
lst_years_start = ['2015', '2016', '2017', '2018', '2019']
# lst_years_start = ['2015', '2016']
lst_years_end = lst_years_start
# lst_years_end = ['2018', '2019']

# lst_years_start = ['2015']
# lst_years_end = ['2019']

# lst_years_start = ['2015']
# lst_years_end = ['2019']
# lst_down_payments = [.1]

lst_fixed_term_mortgage = [5] #2,3,5
int_payments_per_year = 12
lst_loan_life_years = [30] #20, 25, 30

# Stock Calculations Parameters
lst_stocks = ['S&P500', 'Facebook', 'Apple', 'Gold', 'Bitcoin']
# lst_stocks = ['Apple', 'Facebook']

# calculations in total
int_counter_total = 0
int_counter = 0
for float_down_payment in lst_down_payments:
    for str_year_start in lst_years_start:
        for str_year_end in lst_years_end:
            for int_fixed_term_mortgage in lst_fixed_term_mortgage:
                for int_loan_life_year in lst_loan_life_years:
                    for stock in lst_stocks:

                        if int(str_year_start) >= int(str_year_end):
                            continue
                        int_counter_total = int_counter_total + 1


for float_down_payment in lst_down_payments:
    for str_year_start in lst_years_start:
        for str_year_end in lst_years_end:
            for int_fixed_term_mortgage in lst_fixed_term_mortgage:
                for int_loan_life_year in lst_loan_life_years:
                    for stock in lst_stocks:
                        
                        # MORTGAGE CALCULATIONS

                        if int(str_year_start) >= int(str_year_end):
                            continue
                        
                        int_counter = int_counter + 1
                        print('iter', str(int_counter),'of', int_counter_total, ':', float_down_payment, str_year_start, str_year_end, stock)

                        df6 = df5.copy()
                        df6 = df6[df6['REPORT_YEAR']>=str_year_start] #.astype(int)
                        df6 = df6[df6['REPORT_YEAR']<=str_year_end] #.astype(int)

                        # Add a counter by group
    #                     df4 = df4.sort_values(lst_columns_group_by, ascending=[True, True, True, True, True])
                        lst_columns_group_by = ['LEGAL_TYPE', 'ZONE_CATEGORY', 'bedrooms', 'Neighbourhood_Name', 'REPORT_YEAR'] 
                        df6['groupby_counter'] = df6.groupby(lst_columns_group_by[:-1]).cumcount() + 1


                        df6['year_purchased'] = np.where(
                            df6['groupby_counter']==1,
                            df6['REPORT_YEAR'].astype(np.int64),
                            np.nan
                        ) 

#                         print('step 1')

                        df6['year_purchased'] = df6['year_purchased'].ffill()
                        df6['years_since_purchase'] = df6['REPORT_YEAR'].astype(np.int64) - df6['year_purchased']

                        df6['CURRENT_TOTAL_VALUE_CORRECTED'] = np.nan
                        df6['PREVIOUS_TOTAL_VALUE_CORRECTED'] = np.nan
                        for row in range(0, df6.shape[0]):
                            df6['PREVIOUS_TOTAL_VALUE_CORRECTED'].iloc[row] = np.where(
                                df6.groupby_counter.iloc[row] == 1,     
                                df6['PREVIOUS_TOTAL_VALUE'].iloc[row],
                                df6['CURRENT_TOTAL_VALUE_CORRECTED'].iloc[row-1]
                            )

                            df6['CURRENT_TOTAL_VALUE_CORRECTED'].iloc[row] = np.where(
                                df6.groupby_counter.iloc[row] == 1,                            
                                df6['CURRENT_TOTAL_VALUE'].iloc[row],
                                df6['PREVIOUS_TOTAL_VALUE_CORRECTED'].iloc[row] * (1 + (df6['mortgage_return_prctg'].iloc[row] / 100))
                            )
                        df6['PREVIOUS_TOTAL_VALUE_CORRECTED'] = df6['PREVIOUS_TOTAL_VALUE_CORRECTED'].astype(np.int64)
                        df6['CURRENT_TOTAL_VALUE_CORRECTED'] = df6['CURRENT_TOTAL_VALUE_CORRECTED'].astype(np.int64)

#                         print('step 2')

                        df6['owning_opening_costs'] = np.where(
                            df6['groupby_counter']==1, #.astype(int)
                            df6['PREVIOUS_TOTAL_VALUE_CORRECTED'] * float_opening_costs,
                            0)

                        df6['owning_closing_costs'] = np.where(
                            df6['REPORT_YEAR']==str_year_end, #.astype(int)
                            df6['CURRENT_TOTAL_VALUE_CORRECTED'] * float_closing_costs,
                            0)

                        df6['value_at_purchase'] = np.where(
                            df6['groupby_counter']==1,
                            df6['PREVIOUS_TOTAL_VALUE_CORRECTED'],
                            np.nan)
                        df6['value_at_purchase'] = df6['value_at_purchase'].ffill()

                        df6['investment_at_purchase'] = df6['value_at_purchase'] * float_down_payment


#                         print('step 3')
                        # Add mortgage rates
                        df6 = pd.merge(df6, df_average_mortgage_reates,  how='left', left_on=['REPORT_YEAR'], right_on = ['year'])
                        df6.drop(columns=['year', 'month_int'], inplace=True)
                        df6.rename(columns=({'rate': 'mortgage_rate_yearly'}), inplace=True)
                        df6['mortgage_rate_yearly'] = np.where(
                            ((df6['years_since_purchase']==0) | (df6['years_since_purchase']==int_fixed_term_mortgage)),
                            df6['mortgage_rate_yearly'],
                            np.nan
                        )
                        df6['mortgage_rate_yearly'] = df6['mortgage_rate_yearly'].ffill()
                        df6['mortgage_rate_yearly'] = df6['mortgage_rate_yearly'] / 100
                        df6['mortgage_rate_periodical'] = df6['mortgage_rate_yearly'] / int_payments_per_year

                        df6['total_number_payments'] = np.where(
                            ((df6['years_since_purchase']==0) | (df6['years_since_purchase']==int_fixed_term_mortgage)),
                            int_payments_per_year * (int_loan_life_year - df6.years_since_purchase),
                            np.nan
                        )
                        df6['total_number_payments'] = df6['total_number_payments'].ffill()

    #                     df6['mortgage_payments_periodical_so_far'] = int_payments_per_year * (df6.years_since_purchase+1)

#                         print('step 4')

                        df6['mortgage_payment_periodical'] = np.nan
                        df6['mortgage_payment_yearly_principal'] = np.nan
                        df6['mortgage_payment_yearly_interest'] = np.nan
                        df6['loan_amount_at_mortgage_contract'] = np.nan
                        df6['loan_amount_at_purchase'] = np.nan
                        df6['remaining_principal'] = np.nan
                        df6['mortgage_return_value_yearly'] = np.nan

                        for row in range(0, df6.shape[0]):

#                             if row % 500 == 0:
#                                 print('row', row, 'of', df6.shape[0])

                            # at purchase
                            df6['loan_amount_at_mortgage_contract'].iloc[row] = np.where(
                                df6.years_since_purchase.iloc[row] == 0,                            
                                df6['value_at_purchase'].iloc[row] -  df6['investment_at_purchase'].iloc[row],
                                df6['loan_amount_at_mortgage_contract'].iloc[row]
                            )

                            # not at purchase & not up to new mortgage contract
                            df6['loan_amount_at_mortgage_contract'].iloc[row] = np.where(
                                ((df6.years_since_purchase.iloc[row] % lst_fixed_term_mortgage != 0) & (df6.years_since_purchase.iloc[row] != 0)),
                                df6['loan_amount_at_mortgage_contract'].iloc[row-1],
                                df6['loan_amount_at_mortgage_contract'].iloc[row]
                            )

                            # up to new mortgage contract & not at purchase
                            df6['loan_amount_at_mortgage_contract'].iloc[row] = np.where(
                                ((df6.years_since_purchase.iloc[row] % lst_fixed_term_mortgage == 0) & (df6.years_since_purchase.iloc[row] != 0)),
                                df6['remaining_principal'].iloc[row-1],
                                df6['loan_amount_at_mortgage_contract'].iloc[row]
                            )






                            ####### MORTGAGE LIBRARY #######
                            principal_input = df6.loan_amount_at_mortgage_contract.iloc[row]
                            interest_input = df6.mortgage_rate_yearly.iloc[row]
                            term_input = int(df6.total_number_payments.iloc[row] / 12)

                            loan = Loan(principal=principal_input, interest=interest_input, term=term_input)

                            payment_periodical = float(loan.monthly_payment)
                            month_number = int((df6.years_since_purchase.iloc[row] * int_payments_per_year) + 6)

                            interest, principal = (loan.split_payment(number=month_number, amount=Decimal(payment_periodical)))


                            principal = int(principal)
                            interest = int(interest)
                            ###############################

                            ####### NUMPY LIBRARY ##########
    #                         principal_input = df6.loan_amount_at_mortgage_contract.iloc[row]
    #                         interest_input = df6.mortgage_rate_yearly.iloc[row] / 12
    #                         term_input = int(df6.total_number_payments.iloc[row])
    #                         month_number = int((df6.years_since_purchase.iloc[row] * int_payments_per_year) + 6)

    # #                         periodical_total_pmt = abs(np.pmt(interest_input, 360, 400000))
    #                         periodical_principal_pmt = abs(np.ppmt(interest_input, month_number, term_input, principal_input)) #pv[, fv, when]) #	Compute the payment against loan principal.
    #                         periodical_interest_pmt = abs(np.ipmt(interest_input, month_number, term_input, principal_input))
                            ##############################


                            df6['mortgage_payment_yearly_principal'].iloc[row] = principal * 12 # mortgage library
    #                         df6['mortgage_payment_yearly_principal'].iloc[row] = periodical_principal_pmt * 12 # numpy
                            df6['mortgage_payment_yearly_interest'].iloc[row] = interest * 12  # mortgage library
    #                         df6['mortgage_payment_yearly_interest'].iloc[row] = periodical_interest_pmt * 12 # numpy
                            df6['remaining_principal'].iloc[row] = np.where(
                                df6['years_since_purchase'].iloc[row]==0,
                                df6['loan_amount_at_mortgage_contract'].iloc[row] - (principal * 12), # mortgage library
    #                             df6['loan_amount_at_mortgage_contract'].iloc[row] - (periodical_principal_pmt * 12), # numpy
                                df6['remaining_principal'].iloc[row],
                            )

                            df6['remaining_principal'].iloc[row] = np.where(
                                df6['years_since_purchase'].iloc[row]!=0,
                                df6['remaining_principal'].iloc[row-1] - (principal * 12), # mortgage library
    #                             df6['remaining_principal'].iloc[row-1] - (periodical_principal_pmt * 12), # numpy
                                df6['remaining_principal'].iloc[row],
                            )

                        df6['mortgage_payment_yearly_total'] = df6['mortgage_payment_yearly_principal'] + df6['mortgage_payment_yearly_interest']
                        df6['owning_fees_total'] = df6['owning_opening_costs'] + df6['owning_closing_costs'] + df6['total_owners_fees']

                        df6['loan_amount_at_purchase'] = np.where(
                            df6['groupby_counter']==1,
                            df6['loan_amount_at_mortgage_contract'],
                            np.nan
                        )
                        df6['loan_amount_at_purchase'] = df6['loan_amount_at_purchase'].ffill()

#                         print('step 5')

                        df6['mortgage_payment_yearly_total_so_far'] = np.nan
                        df6['owning_fees_total_so_far'] = np.nan
                        for row in range(0, df6.shape[0]):
                            df6['mortgage_payment_yearly_total_so_far'].iloc[row] = np.where(
                            df6['years_since_purchase'].iloc[row]==0,
                            df6['mortgage_payment_yearly_total'].iloc[row],
                            df6['mortgage_payment_yearly_total'].iloc[row] + df6['mortgage_payment_yearly_total_so_far'].iloc[row-1]     
                            )

                            df6['owning_fees_total_so_far'].iloc[row] = np.where(
                            df6['years_since_purchase'].iloc[row]==0,
                            df6['owning_fees_total'].iloc[row],
                            df6['owning_fees_total'].iloc[row] + df6['owning_fees_total_so_far'].iloc[row-1]     
                            )

                        df6['owning_costs_total_yearly'] = df6['mortgage_payment_yearly_total'] + df6['owning_fees_total']
                        df6['owning_costs_total_so_far'] = df6['mortgage_payment_yearly_total_so_far'] + df6['owning_fees_total_so_far']

                        df6['mortgage_investment_value'] = df6['CURRENT_TOTAL_VALUE_CORRECTED'] - df6['owning_costs_total_so_far'] - df6['remaining_principal']

#                         print('step 6')

                        df6['mortgage_return_prctg_yearly'] = np.nan
                        df6['mortgage_return_prctg_cumulative'] = np.nan
                        for row in range(0, df6.shape[0]):
                            
                            df6['mortgage_return_value_yearly'].iloc[row] = np.where(
                                df6['years_since_purchase'].iloc[row]==0,
                                (df6['mortgage_investment_value'].iloc[row] - df6['investment_at_purchase'].iloc[row]),
                                (df6['mortgage_investment_value'].iloc[row] - df6['mortgage_investment_value'].iloc[row-1])
                            )
           
                            df6['mortgage_return_prctg_yearly'].iloc[row] = np.where(
                                df6['years_since_purchase'].iloc[row]==0,
                                (df6['mortgage_return_value_yearly'].iloc[row]) / (df6['investment_at_purchase'].iloc[row]),
                                (df6['mortgage_return_value_yearly'].iloc[row]) / abs(df6['mortgage_investment_value'].iloc[row-1])
                            )
        
                        df6['mortgage_return_prctg_yearly'] = df6['mortgage_return_prctg_yearly'] * 100
                
                        df6['mortgage_return_value_cumulative'] = ((df6['mortgage_investment_value'] - df6['investment_at_purchase']))
                        df6['mortgage_return_prctg_cumulative'] = ((df6['mortgage_investment_value'] - df6['investment_at_purchase']) / df6['investment_at_purchase']) * 100

                        # ((P + G) / P) ^ (1 / n) - 1
                        
                        df6['mortgage_investment_return_cumulative'] = df6['mortgage_return_value_cumulative'] / df6['investment_at_purchase']
#                         df6['mortgage_return_prctg_annualized'] = (((df6['investment_at_purchase'] +  df6['mortgage_return_value_cumulative']) / df6['investment_at_purchase'])**(1 / (df6['years_since_purchase']+1)) - 1) * 100
#                         df6['mortgage_return_prctg_annualized'] = ((df6['investment_at_purchase'] +  df6['mortgage_investment_return_cumulative'])**(1 / (df6['years_since_purchase']+1)) - 1) * 100

                        # needed to fix the formula for negative returns after year 1
                        df6['mortgage_return_prctg_annualized'] = np.where(
                            df6['mortgage_investment_return_cumulative'] > -1,
                            ((1 +  df6['mortgage_investment_return_cumulative'])**(1 / (df6['years_since_purchase']+1)) - 1) * 100,
                            ((1 +  df6['mortgage_investment_return_cumulative'])**(1 / 1) - 1) * 100
                        )
                        
                        
                        df6['year_start'] = str(str_year_start)
                        df6['year_end'] = str(str_year_end)
                        df6['down_payment'] = np.where(
                            float_down_payment==.9999, 
                            "100",
                            str(float_down_payment*100)
                        )
                        df6['Mortgage_Fixed_Term'] = str(int_fixed_term_mortgage)
                        df6['Loan_Life'] = str(int_loan_life_year)
                        df6['stock'] = str(stock)

#                         print('step 7')

                        df6['total_renters_cost_so_far'] = np.nan

                        for row in range(0, df6.shape[0]):

                            df6['total_renters_cost_so_far'].iloc[row] = np.where(
                                df6.groupby_counter.iloc[row] == 1,  
                                df6['total_renters_cost'].iloc[row],
                                df6['total_renters_cost_so_far'].iloc[row-1] + df6['total_renters_cost'].iloc[row]
                            )

#                         print('step 8')
                        
                        
                        ########
                        
                        # STOCK MARGIN CALCULATIONS
                        
                        df6['stock'] = stock
        
                        df6 = pd.merge(df6, df_stock_returns,  how='left', left_on=['REPORT_YEAR', 'stock'], right_on = ['year', 'stock'])

                        # new columns WITH margin
                        df6['investments_stocks_at_start_of_year_w_margin'] = np.nan
                        df6['margin_loan'] = np.nan
                        df6['total_stock_investment_w_margin'] = np.nan
                        df6['stock_return_value_w_margin'] = np.nan
                        df6['margin_interest_value'] = np.nan
                        df6['stocks_w_margin_investment_value_at_end_of_year'] = np.nan
                        df6['return_prctg_yearly_stocks_w_margin'] = np.nan
                        df6['return_value_yearly_stocks_w_margin'] = np.nan

                        # new columns WITHOUT margin
                        df6['investments_stocks_at_start_of_year_wo_margin'] = np.nan
                        df6['total_stock_investment_wo_margin'] = np.nan
                        df6['stock_return_value_wo_margin'] = np.nan
                        df6['stocks_wo_margin_investment_value_at_end_of_year'] = np.nan
                        df6['return_prctg_yearly_stocks_wo_margin'] = np.nan
                        df6['return_value_yearly_stocks_wo_margin'] = np.nan

                        for row in range(0, df6.shape[0]):

                            # with margin
                            df6['investments_stocks_at_start_of_year_w_margin'].iloc[row] = np.where(
                                df6['groupby_counter'].iloc[row]==1,
                                df6['investment_at_purchase'].iloc[row],
                                df6['stocks_w_margin_investment_value_at_end_of_year'].iloc[row-1]
                            ) 


                            float_margin_requirement = df6['margin_requirement'].iloc[row]
                            investments_stocks_at_start_of_year_w_margin = df6['investments_stocks_at_start_of_year_w_margin'].iloc[row]
                            df6['margin_loan'].iloc[row] = (investments_stocks_at_start_of_year_w_margin / (1 - float_margin_requirement)) * float_margin_requirement
                            df6['total_stock_investment_w_margin'].iloc[row] = investments_stocks_at_start_of_year_w_margin + df6['margin_loan'].iloc[row]

                            df6['stock_return_value_w_margin'].iloc[row] = df6['total_stock_investment_w_margin'].iloc[row] * df6['yearly_stock_return'].iloc[row]
                            df6['margin_interest_value'].iloc[row] = df6['margin_loan'].iloc[row] * df6['margin_interest_rate'].iloc[row]
                            df6['stocks_w_margin_investment_value_at_end_of_year'].iloc[row] = investments_stocks_at_start_of_year_w_margin + df6['stock_return_value_w_margin'].iloc[row] - df6['margin_interest_value'].iloc[row]


                            # without margin
                            df6['investments_stocks_at_start_of_year_wo_margin'].iloc[row] = np.where(
                                df6['groupby_counter'].iloc[row]==1,
                                df6['investment_at_purchase'].iloc[row],
                                df6['stocks_wo_margin_investment_value_at_end_of_year'].iloc[row-1]
                            ) 

                            investments_stocks_at_start_of_year_wo_margin = df6['investments_stocks_at_start_of_year_wo_margin'].iloc[row]
                            df6['total_stock_investment_wo_margin'].iloc[row] = investments_stocks_at_start_of_year_wo_margin

                            df6['stock_return_value_wo_margin'].iloc[row] = df6['total_stock_investment_wo_margin'].iloc[row] * df6['yearly_stock_return'].iloc[row]
                            df6['stocks_wo_margin_investment_value_at_end_of_year'].iloc[row] = investments_stocks_at_start_of_year_wo_margin + df6['stock_return_value_wo_margin'].iloc[row]


                        # RETURNS
                        # with margin
                        df6['stocks_w_margin_investment_value_after_rental_costs'] = df6['stocks_w_margin_investment_value_at_end_of_year'] - df6['total_renters_cost_so_far']

                        for row in range(0, df6.shape[0]):
                            
                            df6['return_value_yearly_stocks_w_margin'].iloc[row] = np.where(
                                df6['groupby_counter'].iloc[row]==1,
                                (df6['stocks_w_margin_investment_value_after_rental_costs'].iloc[row] - df6['investments_stocks_at_start_of_year_w_margin'].iloc[row]),
                                (df6['stocks_w_margin_investment_value_after_rental_costs'].iloc[row] - df6['stocks_w_margin_investment_value_after_rental_costs'].iloc[row-1])
                            )
                            
                            df6['return_prctg_yearly_stocks_w_margin'].iloc[row] = np.where(
                                df6['groupby_counter'].iloc[row]==1,
                                (df6['return_value_yearly_stocks_w_margin'].iloc[row]) / abs(df6['investments_stocks_at_start_of_year_w_margin'].iloc[row]),
                                (df6['return_value_yearly_stocks_w_margin'].iloc[row]) / abs(df6['stocks_w_margin_investment_value_after_rental_costs'].iloc[row-1])
                            )
                        df6['return_prctg_yearly_stocks_w_margin'] = df6['return_prctg_yearly_stocks_w_margin'] * 100

                        df6['return_prctg_cumulative_stocks_w_margin'] = (df6['stocks_w_margin_investment_value_after_rental_costs'] - df6['investment_at_purchase']) / df6['investment_at_purchase']
                        df6['return_prctg_cumulative_stocks_w_margin'] = df6['return_prctg_cumulative_stocks_w_margin'] * 100
                        
                        df6['return_value_cumulative_stocks_w_margin'] = ((df6['stocks_w_margin_investment_value_after_rental_costs'] - df6['investment_at_purchase']))
                        
                        df6['return_rate_total_cumulative_w_margin'] = (df6['investment_at_purchase'] +  df6['return_value_cumulative_stocks_w_margin']) / df6['investment_at_purchase']
                        
#                         df6['return_prctg_cumulative_stocks_w_margin_annualized'] = (((df6['investment_at_purchase'] +  df6['return_value_cumulative_stocks_w_margin']) / df6['investment_at_purchase'])**(1 / (df6['years_since_purchase']+1)) - 1) * 100
                        df6['return_prctg_cumulative_stocks_w_margin_annualized'] = np.where(
                            df6['return_rate_total_cumulative_w_margin'] > -1,
                            ((1 + df6['return_rate_total_cumulative_w_margin']) **(1 / (df6['years_since_purchase']+1)) - 1) * 100,
                            ((1 + df6['return_rate_total_cumulative_w_margin']) **(1 / 1) - 1) * 100
                        )
                        
                        

                        # RETURNS
                        # WITHOUT margin
                        df6['stocks_wo_margin_investment_value_after_rental_costs'] = df6['stocks_wo_margin_investment_value_at_end_of_year'] - df6['total_renters_cost_so_far']

                        for row in range(0, df6.shape[0]):
                            
                            df6['return_value_yearly_stocks_wo_margin'].iloc[row] = np.where(
                                df6['groupby_counter'].iloc[row]==1,
                                (df6['stocks_wo_margin_investment_value_after_rental_costs'].iloc[row] - df6['investments_stocks_at_start_of_year_wo_margin'].iloc[row]),
                                (df6['stocks_wo_margin_investment_value_after_rental_costs'].iloc[row] - df6['stocks_wo_margin_investment_value_after_rental_costs'].iloc[row-1])
                            )
                            
                            df6['return_prctg_yearly_stocks_wo_margin'].iloc[row] = np.where(
                                df6['groupby_counter'].iloc[row]==1,
                                (df6['return_value_yearly_stocks_wo_margin'].iloc[row]) / abs(df6['investments_stocks_at_start_of_year_wo_margin'].iloc[row]),
                                (df6['return_value_yearly_stocks_wo_margin'].iloc[row]) / abs(df6['stocks_wo_margin_investment_value_after_rental_costs'].iloc[row-1])
                            )
                        df6['return_prctg_yearly_stocks_wo_margin'] = df6['return_prctg_yearly_stocks_wo_margin'] * 100

                        df6['return_prctg_cumulative_stocks_wo_margin'] = (df6['stocks_wo_margin_investment_value_after_rental_costs'] - df6['investment_at_purchase']) / df6['investment_at_purchase']
                        df6['return_prctg_cumulative_stocks_wo_margin'] = df6['return_prctg_cumulative_stocks_wo_margin'] * 100

                        df6['return_value_cumulative_stocks_wo_margin'] = ((df6['stocks_wo_margin_investment_value_after_rental_costs'] - df6['investment_at_purchase']))
                        
                        df6['return_rate_total_cumulative_wo_margin'] = (df6['investment_at_purchase'] +  df6['return_value_cumulative_stocks_wo_margin']) / df6['investment_at_purchase']
                        
#                         df6['return_prctg_cumulative_stocks_wo_margin_annualized'] = (((df6['investment_at_purchase'] +  df6['return_value_cumulative_stocks_wo_margin']) / df6['investment_at_purchase'])**(1 / (df6['years_since_purchase']+1)) - 1) * 100
                        df6['return_prctg_cumulative_stocks_wo_margin_annualized'] = np.where(
                            df6['return_rate_total_cumulative_wo_margin'] > -1,
                            ((1 + df6['return_rate_total_cumulative_wo_margin']) **(1 / (df6['years_since_purchase']+1)) - 1) * 100,
                            ((1 + df6['return_rate_total_cumulative_wo_margin']) **(1 / 1) - 1) * 100
                        )
                        
                        
                        #######
                        
#                         print('step 9')



                        str_file = str(str_year_start) + '_' + str(str_year_end) + '_' + str(float_down_payment) + '_' + str(int_fixed_term_mortgage) + '_' + str(int_loan_life_year) + '_' + str(stock)
    #                     df6[lst_columns].to_csv(str_directory + '/csv/' + str_file + '.csv', index=False)
#                         df6[lst_columns].to_excel(lst_directory[int_environment] + 'xlsx/' + str_file + '.xlsx', index=False)
#                         df6.to_excel(lst_directory[int_environment] + 'xlsx/' + str_file + '.xlsx', index=False)
                        df6[lst_columns].to_csv(lst_directory[int_environment] + 'csv/' + str_file + '.csv', index=False)
```
